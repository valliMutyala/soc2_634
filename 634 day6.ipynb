```python
#----->unsupervised learning algorithm
#k means clustering
#k means clustering is an unsupervised learning algorithm that will attempt to group similar clusters together from your data
#it is mainly used in 
#clustering similar ocuments
#clustering customers based on similar features
```

```python
from sklearn.datasets import make_blobs
```

```python
# the make_blobs function is used to generate synthetic datasets
# for cludtering and classification tasks
# the function willcreate clusters of data points with gaussian distributions
```

```python
# creating random dataset
data=make_blobs(n_samples=200,n_features=2,centers=4,cluster_std=1.6,random_state=101)
data
# n_samples=Total number of points equally divided among clusters
# n_features=it indicates no of features(columns)
# centers=it determine no.of clusters to be generated
# cluster_std=it selects the standard deviation of the clusters.high value makes the clusters to spread out
```

```python
import matplotlib.pyplot as plt
x,y=data
plt.scatter(x[:,0],x[:,1],c=y,edgecolor='black',s=100,cmap='rainbow')
plt.xlabel("feature 1")
plt.ylabel("feature 2")
plt.title("scatter plot for k means")
```

```python
#to nmae this columns
df_feat=pd.DataFrame(Scaled_features,columns=df.columns[:-1])
df_feat
```

```python
from sklearn.model_selection import train_test_split
x=df_feat
y=df['TARGET CLASS']
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3)
```

```python
x.shape
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
KNN=KNeighborsClassifier(n_neighbors=3)
```

```python
#to train model
KNN.fit(x_train,y_train)
```

```python
pred=KNN.predict(x_test)
```

```python
pred
```

```python
from sklearn.metrics import accuracy_score
```

```python
acc=accuracy_score(pred,y_test)
```

```python
acc
```

```python
#to find the error rate
error_rate=[]
for val in range(1,30):
    knn=KNeighborsClassifier(n_neighbors=val)
    knn.fit(x_train,y_train)
    pred_i=knn.predict(x_test)
    error_rate.append(np.mean(pred_i !=y_test))
error_rate
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(6,3))
plt.plot(range(1,30),error_rate)
plt.title("Error rate vs value")
plt.xlabel("k")
plt.ylabel("error rate")
```
